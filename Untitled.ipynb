```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load a weather dataset
weather_df = pd.read_csv("https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv")

# Display first few rows
weather_df.head()
```

```python
# Convert 'Date' column to datetime
weather_df['Date'] = pd.to_datetime(weather_df['Date'])
```

```python
# Plot temperature trends

plt.plot(weather_df['Date'], weather_df['Temp'], label = 'Temperature', color = 'Blue')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Daily Minimum Temperature Trends')
plt.legend()
plt.show()
```
